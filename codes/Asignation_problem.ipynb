```python
from pyomo.environ import*
from pyomo.opt import*
model=ConcreteModel()


model.x1a=Var(within=Binary)
model.x1b=Var(within=Binary)
model.x1c=Var(within=Binary)
model.x1d=Var(within=Binary)
model.x2a=Var(within=Binary)
model.x2b=Var(within=Binary)
model.x2c=Var(within=Binary)
model.x2d=Var(within=Binary)
model.x3a=Var(within=Binary)
model.x3b=Var(within=Binary)
model.x3c=Var(within=Binary)
model.x3d=Var(within=Binary)
model.x4a=Var(within=Binary)
model.x4b=Var(within=Binary)
model.x4c=Var(within=Binary)
model.x4d=Var(within=Binary)

model.obj=Objective(expr=5*model.x1a+4*model.x1b+6*model.x1c+5*model.x1d+
                    6*model.x2a+6*model.x2b+7*model.x2c+5*model.x2d+
                    7*model.x3a+5*model.x3b+7*model.x3c+6*model.x3d+
                    5*model.x4a+4*model.x4b+6*model.x4c+6*model.x4d,sense=minimize)

model.con1=Constraint(expr=model.x1a+model.x1b+model.x1c+model.x1d==1)
model.con2=Constraint(expr=model.x2a+model.x2b+model.x2c+model.x2d==1)
model.con3=Constraint(expr=model.x3a+model.x3b+model.x3c+model.x3d==1)
model.con4=Constraint(expr=model.x4a+model.x4b+model.x4c+model.x4d==1)

model.con5=Constraint(expr=model.x1a+model.x2a+model.x3a+model.x4a==1)
model.con6=Constraint(expr=model.x1b+model.x2b+model.x3b+model.x4b==1)
model.con7=Constraint(expr=model.x1c+model.x2c+model.x3c+model.x4c==1)
model.con8=Constraint(expr=model.x1d+model.x2d+model.x3d+model.x4d==1)

solver=SolverFactory('glpk')
results=solver.solve(model)
print(results)


print('The value of x1A is', value(model.x1a))
print('The value of x1B is', value(model.x1b))
print('The value of x1C is', value(model.x1c))
print('The value of x1D is', value(model.x1d))
print('The value of x2A is', value(model.x2a))
print('The value of x2B is', value(model.x2b))
print('The value of x2C is', value(model.x2c))
print('The value of x2D is', value(model.x2d))
print('The value of x3A is', value(model.x3a))
print('The value of x3B is', value(model.x3b))
print('The value of x3C is', value(model.x3c))
print('The value of x3D is', value(model.x3d))
print('The value of x4A is', value(model.x4a))
print('The value of x4B is', value(model.x4b))
print('The value of x4C is', value(model.x4c))
print('The value of x4D is', value(model.x4d))










```
